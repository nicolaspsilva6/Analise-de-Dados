#Preparaçao de Dados
Solucinar os problemas identificados


```python
df.columns
```

```python
colunas=['matricula_funcionario', 'nome_Funcionario', 'cargo', 'codigo_loja',
       'nome_loja', 'codigo_produto', 'descricao_produto', 'categoria',
       'preco_custo', 'valor_unitario', 'quantidade', 'comissao', 'dt_venda',
       'dt_entrega']
```

```python
#CONVERTENDO objetc PARA STRING
df["matricula_funcionario"]=df["matricula_funcionario"].astype(str)
```

```python
#tirar virgula e colocar em ponto
df['valor_unitario']=df['valor_unitario'].str.replace(',','.')
df['preco_custo']=df['preco_custo'].str.replace(',','.')
df['comissao']=df['comissao'].str.replace(',','.')

```

```python
#convertendo string em float- preço_custo e valor_unitario e _comissao
df = df.astype(
    {
     'valor_unitario':np.float64,
     'preco_custo':np.float64,
     'comissao':np.float64
    }
)





```

```python
df['dt_entrega']=pd.to_datetime(df['dt_entrega'], format='%d/%m/%Y %H:%M')
df['dt_entrega']
```

```python
#indentificar e tratar valores faltantes (missing valoue)
df.isnull().sum()
```

```python
#metodo fillna
#adicionar valor aonde esta em 0
df['comissao'].fillna(0)
```

```python
#metodo ffill()
#encontra o valor e da continudade ao que foi encontrado
df.comissao.ffill()
```

```python
#o mesmo metodo do ffill porem ele vai pra cima
df['comissao'].bfill().head(50)
```

```python
#apagar as linhas com valores 0
df.dropna()
```

```python
#uma forma um pouco melhor e subtituir os dados pelo valores faltantes pelo valor médio de variavel.
df['comissao'].mean()
```

```python
#adicionado o valor aonde esyava zerado
df['comissao'].fillna(df['comissao'].mean())
```

```python
#criando coluno no Data Frame
df['valor_total']=df["quantidade"]*df["valor_unitario"]
df.head()
```

```python
#ordenar numeros em ordem do menor pro maior ou do menor para o maior
df.sort_values(["Categoria","valor_total"],ascending=[True,False],inplace=True)
```
